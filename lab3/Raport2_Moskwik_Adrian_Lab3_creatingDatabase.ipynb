```python
Base = declarative_base() 
from sqlalchemy import Column, Integer, String, Date
from sqlalchemy import ForeignKey

class Cases(Base):
    __tablename__ = 'cases'
    id = Column(Integer, primary_key=True)
    suspected_cases = Column(Integer)
    probable_cases = Column(Integer)
    confirmed_cases = Column(Integer)
    all_cases = Column(Integer)
                       
    def __repr__(self):
        return "<cases(id='{0}', suspected_cases={1}, probable_cases={2}, confirmed_cases={3}, all_cases={4})>".format(
            self.id, self.suspected_cases, self.probable_cases, self.confirmed_cases, self.all_cases)
                       
class Deaths(Base):
    __tablename__ = 'deaths'
    id = Column(Integer, primary_key=True)
    suspected_deaths = Column(Integer)
    probable_deaths = Column(Integer)
    confirmed_deaths = Column(Integer)
    all_deaths = Column(Integer)
                       
    def __repr__(self):
        return "<deaths(id='{0}', suspected_deaths={1}, probable_deaths={2}, confirmed_deaths={3}, all_deaths={4})>".format(
            self.id, self.suspected_deaths, self.probable_deaths, self.confirmed_deaths, self.all_deaths)      

                        
class Country(Base):
    __tablename__ = 'country'
    id = Column(Integer, primary_key=True)
    name = Column(String(50))
                        
    def __repr__(self):
        return "<country(id='{0}',name={1})>".format(self.id,self.name)

                        
class Ebola_2014_2016(Base):
    __tablename__ = 'ebola_2014_2016'
    id = Column(Integer,primary_key=True)
    id_country = Column(Integer, ForeignKey('country.id'))
    date = Column(Date)
    id_cases = Column(Integer, ForeignKey('cases.id'))
    id_deaths = Column(Integer, ForeignKey('deaths.id'))
                        
    def __repr__(self):
        return "<ebola_2014_2016(id='{0}', id_country={1}, date={2}, id_cases={3}, id_deaths={4})>".format(
            self.id, self.id_country, self.date, self.id_cases, self.id_deaths)      
```
